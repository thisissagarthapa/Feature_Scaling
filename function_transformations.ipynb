```python
np.random.seed(42)  # For reproducibility
dataset['new_outlier_column'] = np.random.normal(loc=50, scale=10, size=3353)  # Normal distribution

# Adding outliers: Inserting some extreme values (e.g., 100, 120) in random places
outlier_indices = np.random.choice(dataset.index, size=10, replace=False)  # Randomly select 10 rows
dataset.loc[outlier_indices, 'new_outlier_column'] = np.random.choice([100, 120, 150], size=10)

# Check the first few rows of the updated DataFrame
print(dataset.head())
```

```python
dataset.describe()
```

```python
sns.boxplot(x="new_outlier_column",data=dataset)
plt.show()
```

```python
sns.distplot(dataset["new_outlier_column"])
plt.show()
```

```python
q1=dataset["new_outlier_column"].quantile(0.25)
q3=dataset["new_outlier_column"].quantile(0.75)

q1,q3
```

```python
iqr=q3-q1
iqr
```

```python
min_r=q1-(1.*iqr)
max_r=q3+(1.*iqr)
min_r,max_r
```

```python
dataset["updated_column"] = dataset["new_outlier_column"][(dataset["new_outlier_column"] <= max_r) & (dataset["new_outlier_column"] >= min_r)]
```

```python
sns.distplot(dataset["updated_column"])
plt.show()
```

```python
from sklearn.preprocessing import FunctionTransformer
```

```python
ft=FunctionTransformer(func=np.log1p)
```

```python
dataset["updated_column_tf"]= ft.fit_transform(dataset[["updated_column"]])
```

```python
plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
sns.distplot(dataset["updated_column"])
plt.title("BEFORE")

plt.subplot(1,2,2)
sns.distplot(dataset["updated_column_tf"])
plt.title("AFTER")
plt.show()
```
